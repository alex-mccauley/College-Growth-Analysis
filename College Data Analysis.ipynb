```python
def get_attendance_data_year(year):
    efa = pd.read_csv('Enrollment Data\ef%da.csv' % year)
    y = str(year)
    
    d = {'EFTOTLT':'total_' + y,\
         'EFALEVEL':'class'}
    efa = efa[['UNITID','EFALEVEL','EFTOTLT']]
    efa.rename(columns=d,inplace=True)
    rnm = {22:'ug', 24:'ug_freshmen', 32:'graduate'}
    
    def map_col(val):
        if val in rnm.keys():
            return rnm[val]
        else:
            return np.nan
    efa['class'] = efa['class'].apply(map_col)
    efa.dropna(axis=0,subset=['class'],inplace=True) # drop other rows
    
    # set up for layered columns; 
    col1 = ['UNITID', 'class' ] + [y]*len(d.keys())
    col2 = efa.columns
       
    return efa    
```

```python
def get_attendence_data(year_list):
    total_data = []
    for year in year_list:
        df = get_attendance_data_year(year)
        if len(total_data) == 0:
            total_data = df
        else:            
            total_data = pd.merge(total_data, df,on=['UNITID','class'])
    return total_data
```

```python
# simple helper functions
def get_group(df,key,val, negation=False):
    if negation:
        return df[df[key] != val]
    else:
        return df[df[key] == val]

# chain together for multiple selection - pass dictionary of key:value pairs
def get_groups(df, tups):
    df2 = df
    for tup in tups:
        key = tup[0]
        val = tup[1]
        if len(tup)>2:
            negation = tup[2]
        else:
            negation = False
        df2 = get_group(df2, key, val, negation=negation)
    return df2
```

### Extract yearly time series for total students

```python
def get_time_series(df_in, year_list, get_val = 'total'):
    
    df = pd.DataFrame(df_in[['UNITID','class']])
    
    for year in year_list:
        col = get_val + '_' + str(year)        
        df[str(year)] = df_in[col]
    return df
```

```python
def compute_linear_fit(df, year_list, fit_val='total'):
    ts = get_time_series(df, year_list, get_val=fit_val)
    
    Y = ts.drop(axis=1,labels=['UNITID','class']).values.transpose() # shaped nsamples x nfeatures - each college is a "feature"
    x = np.array(year_list)
    
    w = np.polyfit(x,Y,1,full=True)
    
    # read into dataframe
    df_out = pd.DataFrame()
    df_out['UNITID'] = ts['UNITID']
    df_out['class'] = ts['class']
    
    df_out['mean'] = np.mean(Y,axis=0)
    df_out['growth'] = w[0][0]
    df_out['intercept'] = w[0][1]
    df_out['percent_growth'] = df_out['growth'] / df_out['mean']
    df_out['residual'] = np.sqrt(w[1]) / (df_out['mean'] * np.sqrt(len(year_list)))
    
    return df_out                                          
```

```python
growth_4yr = compute_linear_fit(enrollments_locations_4yr, years, fit_val='total')
growth_4yr = pd.merge(left=enrollments_locations_4yr,right=growth_4yr,on=['UNITID','class'],how='inner')
```

Peform linear fit on this dataset - tricky part is time series are of unequal lengths

```python
cols = median_rental_price_sqft_2.drop(labels=['RegionName','City','State','Metro','CountyName'],axis=1).columns
periods = map(lambda x:pd.Period(x), cols.values)
```

```python
def get_rents(row):
    Yi = row.drop(labels=['zip_code','City','State','Metro','CountyName']).dropna().values
    return map(np.float, Yi)

def get_fit_simple(row):
    # extract out (yearly) growth numbers
    Y = get_rents(row)
    x = range(len(Y))
    w = np.polyfit(x,Y,1,full=False)
    return 12.0 * w[0] # extrapolate to yearly rent/month growth

def get_last_rent(row):
    Y = get_rents(row)
    return np.mean(Y)

median_rental_price_sqft_growth = median_rental_price_sqft_2.copy()
median_rental_price_sqft_growth['yearly_rent_growth_sqft'] = median_rental_price_sqft_growth.apply(get_fit_simple,axis=1)
median_rental_price_sqft_growth['monthly_rent_sqft'] = median_rental_price_sqft_growth.apply(get_last_rent,axis=1)
median_rental_price_sqft_growth['rent_growth_rate'] = median_rental_price_sqft_growth['yearly_rent_growth_sqft']/median_rental_price_sqft_growth['monthly_rent_sqft']

# ensure that zip codes have 5 digits
median_rental_price_sqft_growth['zip_code'] = median_rental_price_sqft_growth.apply(lambda x:str(x['zip_code']).zfill(5),axis=1)
```

Define function to compute distances between two zip codes

```python
import math
 
def distance_on_unit_sphere(lat1, long1, lat2, long2):
 
    try:
        # Convert latitude and longitude to 
        # spherical coordinates in radians.
        degrees_to_radians = math.pi/180.0

        # phi = 90 - latitude
        phi1 = (90.0 - lat1)*degrees_to_radians
        phi2 = (90.0 - lat2)*degrees_to_radians

        # theta = longitude
        theta1 = long1*degrees_to_radians
        theta2 = long2*degrees_to_radians

        # Compute spherical distance from spherical coordinates.

        # For two locations in spherical coordinates 
        # (1, theta, phi) and (1, theta, phi)
        # cosine( arc length ) = 
        #    sin phi sin phi' cos(theta-theta') + cos phi cos phi'
        # distance = rho * arc length

        cos = (math.sin(phi1)*math.sin(phi2)*math.cos(theta1 - theta2) + 
               math.cos(phi1)*math.cos(phi2))
        arc = math.acos( cos )

        # Remember to multiply arc by the radius of the earth 
        # in your favorite set of units to get length.
        return 3600.0*arc
    except:
        return 1.0e6 # punt on correcting this error

# compute distances between zip1 and all the zips in rental_df, using the lat/long data in zip_df
def distance_to_zip(zip1, rental_df, zip_df):
    
    lat1 = zip_df.ix[zip1].latitude
    long1 = zip_df.ix[zip1].longitude
    
    df2 = rental_df.copy()
    df2['distance'] = df2.apply(lambda row:distance_on_unit_sphere(lat1,long1,row['latitude'],row['longitude']),axis=1)
    
    df3 = pd.DataFrame(df2[['zip_code','distance']],columns=['zip_code','distance'])
    
    # pick top 10 closest zip codes, then find the zip that is numerically closest
    df3 = df3.sort('distance',ascending=True).iloc[:100]
    
    # sort by zip distance
    #df3['zip_rel'] = np.abs(df3['zip_code'].apply(lambda x:float(x[1:]) - float(zip1[1:])))
    df3['zip_rel'] = np.abs(df3['zip_code'].apply(lambda x:float(x) - float(zip1)))
    return df3.sort('zip_rel',ascending=True)

def get_nearest_rental_data(zip1, rental_df, zip_df):
    df = distance_to_zip(zip1, rental_df, zip_df)
    nearest_zip = df.iloc[0]['zip_code']
    row = rental_df.set_index('zip_code').ix[nearest_zip]
    row['zip_code'] = zip1
    row['actual_zip_code'] = nearest_zip # zip where we got the actual rental data
    return row
```

### Add lat/long to college address data

Get lat/log of addresses for easy mapping;
source: https://pypi.python.org/pypi/geopy/1.9.1

```python
from geopy.geocoders import Nominatim

def get_address(row):
    return (row['address'],row['city'],row['state'],row['zip_code'])

def get_lat_long(street,city,state,zip_code):
    #try exact address
    geolocator = Nominatim()
    loc = "%s, %s, %s, %s"%(street,city,state,zip_code)
    
    try:
        location = geolocator.geocode(loc)
    except: # timeout
        # try again without exact street
        print "timeout on address %s, retrying full address" % loc
        try:
            location = geolocator.geocode(loc)
        except:
            print "second timeout on address %s, trying partial address" % loc
            loc = "%s, %s, %s"%(city,state,zip_code)
            location = geolocator.geocode(loc)
        
    if location is None:
        # try again without exact street
        loc = "%s, %s, %s"%(city,state,zip_code)
        location = geolocator.geocode(loc)
        
    if location is None:        
        return 'none_found'
    
    return (location.latitude, location.longitude)
```

```python
def get_lat_long_from_uid(d_lat, uid, key):
    if uid in d_lat.keys():
        ll = d_lat[uid]        
        if ll == 'timeout' or ll == 'none_found':
            return 'timeout'
        else:
            if key == 'lat':
                return ll[0]
            else:
                return ll[1]
    else:
        return None
    
def update_total_data(df, lat_long_dict):
    # update df lat/long entries
    for uid in lat_long_dict.keys():
        df.ix[df['UNITID'] == uid,'lat'] = get_lat_long_from_uid(lat_long_dict, uid, 'lat')
        df.ix[df['UNITID'] == uid,'long'] = get_lat_long_from_uid(lat_long_dict, uid, 'long')
    return df
```

```python
# find all non-populated lat/long entries
# sort by class size
# load lat/long data
lat_long_file = 'lag_long_dict.txt'
try:
    f = open(lat_long_file,'rb')
    d_lat_long = pickle.load(f)
    f.close()
except:
    d_lat_long = defaultdict(dict)

# modify total_data first in case it was re-read
total_data = update_total_data(total_data, d_lat_long) # modifies total_data in-place

ug_total = total_data.groupby('class').get_group('ug')
#ug_total.set_index(keys='UNITID',inplace=True)

# find all unitids where lat/long = None
ug_total = ug_total[ug_total['lat'].map(lambda x:x is None)]

# sort by largest class
ug_total.sort(axis=0,columns='Total',ascending=False,inplace=True)

print "%g lat/long entries to calculate" % ug_total.shape[0]

# calculate lat/long for a block
block_size = 20
for idx in ug_total.index[:block_size]:
    add = ug_total.ix[idx,['UNITID', 'address','city','state','zip_code']]    
    try:
        ll = get_lat_long(add.address,add.city,add.state,add.zip_code.split('-')[0])    
    except:
        print "timeout on address: %s, %s, %s, %s" % (add.address,add.city,add.state,add.zip_code)
        d_lat_long[add.UNITID] = 'timeout'
        continue
    
    if ll == 'none_found':
        print "unable to locate address: %s, %s, %s, %s" % (add.address,add.city,add.state,add.zip_code)
        d_lat_long[add.UNITID] = 'none_found'
    d_lat_long[add.UNITID] = ll

# save lat/long data
f = open(lat_long_file,'wb')
pickle.dump(d_lat_long,f)
f.close()

# write into total data as well
total_data = update_total_data(total_data, d_lat_long) # modifies total_data in-place
```
